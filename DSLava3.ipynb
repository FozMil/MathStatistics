```python
import numpy as np
import scipy.stats as stats

# Генерация случайных значений для испытаний Бернулли
n = 1000  # количество испытаний
p = 0.5  # вероятность успеха в каждом испытании
alpha = 0.05 # Уровень значимости
bernoulli_trials = np.random.binomial(1, p, n)  # генерация случайных значений (0 или 1) для n испытаний

# Вывод результатов в консоль
successes = np.sum(bernoulli_trials)
failures = n - successes
print("Успешных испытаний:", successes)
# Проверка гипотезы с использованием критических значений
# Вычисление критических значений

# Вычисление наблюдаемого значения статистики критерия
g_observed = successes

# Вычисление критических значений
g_critical1 = n*p-(alpha/2)*n
g_critical2 = n*p+(alpha/2)*n

# Проверка гипотезы с использованием критических значений
if g_observed <= g_critical1 or g_observed >= g_critical2:
    print("Гипотеза H1 принимается")
else:
    print("Гипотеза H0 принимается")

# Вычисление p-значения
p_value = 2 * min(np.fabs(g_critical2-g_observed)/(n*p), np.fabs(g_observed-g_critical1)/(n*p))

# Проверка гипотезы с использованием p-значения
if p_value > alpha:
    print("Гипотеза H1 принимается")
else:
    print("Гипотеза H0 принимается")
```

```python
import scipy.stats as stats

def test_binomial_hypothesis(observed_successes, total_trials, hypothesized_prob, alpha, alternative):
    # Вычисляем наблюдаемую долю успехов
    observed_prob = observed_successes / total_trials
    
    # 1. Способ: Сравнение с критическими значениями
    if alternative == 'less':
        critical_value = stats.binom.ppf(alpha, total_trials, hypothesized_prob)
        reject_null = observed_successes <= critical_value
    elif alternative == 'greater':
        critical_value = stats.binom.ppf(1 - alpha, total_trials, hypothesized_prob)
        reject_null = observed_successes >= critical_value
    elif alternative == 'two-sided':
        critical_value1 = stats.binom.ppf(alpha / 2, total_trials, hypothesized_prob)
        critical_value2 = stats.binom.ppf(1 - alpha / 2, total_trials, hypothesized_prob)
        reject_null = observed_successes <= critical_value1 or observed_successes >= critical_value2

    # 2. Способ: Использование P-значения
    if alternative == 'less':
        p_value = stats.binom.cdf(observed_successes, total_trials, hypothesized_prob)
    elif alternative == 'greater':
        p_value = 1 - stats.binom.cdf(observed_successes - 1, total_trials, hypothesized_prob)
    elif alternative == 'two-sided':
        p_value = 2 * min(stats.binom.cdf(observed_successes, total_trials, hypothesized_prob),
                         1 - stats.binom.cdf(observed_successes - 1, total_trials, hypothesized_prob))

    # Принятие решения
    if reject_null:
        print(f"Reject H0 in favor of H1 at {alpha} significance level")
    else:
        print(f"Fail to reject H0 at {alpha} significance level")

    print(f"P-value: {p_value}")

# Пример использования:
observed_successes = 8
total_trials = 10
hypothesized_prob = 0.5
alpha = 0.025
alternative = 'two-sided'

test_binomial_hypothesis(observed_successes, total_trials, hypothesized_prob, alpha, alternative)
```
