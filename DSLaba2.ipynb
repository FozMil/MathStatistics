```python
import numpy as np
from scipy.stats import norm
from scipy.stats import chi2

# Функция для расчета доверительного интервала с использованием точного метода
def confidence_interval_exact(data, alpha):
    n = len(data)
    mean = np.mean(data)
    std_dev = np.var(data, ddof=1)
    t = norm.ppf((1 + alpha) / 2)
    chi2_lower = chi2.ppf((alpha / 2, df=n-1)
    lower_bound = mean - t * std_dev / np.sqrt(n)
    upper_bound = mean + t * std_dev / np.sqrt(n)
    return lower_bound, upper_bound

# Функция для расчета доверительного интервала с использованием метода ЦПТ
def confidence_interval_approx(data, alpha):
    n = len(data)
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)
    z = norm.ppf(1 - (1 - alpha) / 2)
    lower_bound = mean - z * std_dev / np.sqrt(n)
    upper_bound = mean + z * std_dev / np.sqrt(n)
    return lower_bound, upper_bound
# Функция для расчета доверительного интервала для дисперсии с использованием точного метода
#def confidence_interval_approx(data, alpha):
#    n = len(data)
#    chi2_lower = chi2.ppf(alpha / 2, df=n-1)
#    chi2_upper = chi2.ppf(1 - (alpha) / 2, df=n-1)
#    lower_bound = (n-1) * np.var(data, ddof=1) / chi2_lower
#    upper_bound = (n-1) * np.var(data, ddof=1) / chi2_upper
#    return lower_bound, upper_bound

# Заданные параметры
N_values = [20, 50, 100]
alpha_values = [0.9, 0.95, 0.99]

# Генерация выборок и расчет доверительных интервалов
for N in N_values:
    for alpha in alpha_values:
        data = np.random.normal(loc=0, scale=1, size=N)  # Генерация выборки из нормального распределения
        exact_interval = confidence_interval_exact(data, alpha)
        approx_interval = confidence_interval_approx(data, alpha)
        print(f"N={N}, alpha={alpha}: точный интервал {exact_interval}, приближенный интервал {approx_interval}")
```

```python
import numpy as np
import pandas as pd
import statistics as st
from scipy import stats
from tabulate import tabulate
from scipy.stats import norm
from scipy.stats import chi2


def s_2(xs, n):
    mu = np.mean(xs)
    square_deviation = lambda x : (x - mu) ** 2 
    return sum( map(square_deviation, xs) ) / (n - 1)
    
# Функция для расчета доверительного интервала с использованием точного метода
def interv_ex(data, alpha, N):
    n = N
    #print(data)
    s2 = s_2(data, N)
    chi_l = chi2.ppf((1+alpha)/2, df=n-1)
    chi_u = chi2.ppf((1-alpha)/2, df=n-1)
    left = (n-1) * s2 / chi_l
    right = (n-1) * s2 / chi_u
    return left, right
# Функция для расчета доверительного интервала с использованием приближенного метода
def interv_apr(data, alpha, N):
    n = N
    s2 = s_2(data, N)
    f = norm.ppf((1+alpha)/2)
    left = s2 - s2 * np.sqrt(2/(n-1)) * f
    right =  s2 + s2 * np.sqrt(2/(n-1)) * f
    return left, right

N_values = [20, 50, 100]
alpha_values = [0.9, 0.95, 0.99]
D_value = 20
E_value = 10
print(f"Теоретическое значение дисперсии = {D_value/50}")
#sample = stats.norm(loc = 0, scale = np.sqrt(D)).rvs(N_values[2])
#sample = stats.uniform(-0.05, 2).rvs(N_values[0])
#sample[:N_values[0]]
#print(statistics.NormalDist.from_samples(sample))
#print(statistics.NormalDist.from_samples(sample).variance)
sample = st.NormalDist(E_value, np.sqrt(D_value/50))
left, right = interv_ex(sample.samples(N_values[0], seed = 10), alpha_values[0], N_values[0])  
print(f"N={N_values[0]}\talpha={alpha_values[0]}\t точный интервал {left},{right}\tширина={right-left}")
leftA, rightA = interv_apr(sample.samples(N_values[0], seed = 10), alpha_values[0], N_values[0])
print(f"N={N_values[0]}\talpha={alpha_values[0]}\t приближенный интервал {leftA},{rightA}\tширина={rightA-leftA}")
```
