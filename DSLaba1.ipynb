```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy. stats import norm

np.random.seed (0)

x = np.random.normal (10, 1, size=20)
print (x)
var = np.var(x, ddof = 1)
print (var)

sns.displot (x, kde=True)

#plt.legend(title='Parameters')

plt.ylabel('Density')
plt.xlabel('x')
plt.title('Normal Distributions', fontsize= 14 )
```

```python
import numpy as np
import statistics
import matplotlib.pyplot as plt
from scipy. stats import norm

x = np.arange(-15, 45, 0.001)

plt.plot (x, norm. pdf (x, 1, 1), label=': 1, : 1', color='gold')
plt.plot (x, norm. pdf (x, 10, 1), label=': 10, : 1', color='red')
plt.plot (x, norm. pdf (x, 20, 1), label=': 20, : 1', color='pink')
print(statistics.NormalDist.from_samples(x))
print(x)

plt.ylabel('Density')
plt.xlabel('x')
plt.legend()
plt.title('Normal Distributions', fontsize= 14 )
```
